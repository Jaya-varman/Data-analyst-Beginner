```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

myfile = pd.read_csv("temperatures.csv")

myfile = myfile.dropna(subset=['ANNUAL'])

#  Fastest Warming Months 
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
          'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

slopes = {}
for month in months:
    z = np.polyfit(myfile['YEAR'], myfile[month], 1)
    slopes[month] = z[0]  # Slope of the line

# Sort and display top 3 warming months
sorted_slopes = sorted(slopes.items(), key=lambda x: x[1], reverse=True)
top_3 = sorted_slopes[:3]

print(" Fastest Warming Months:")
for month, slope in top_3:
    print(f"  - {month}: {slope:.4f} Â°C/YEAR")

# Hottest & Coldest YEARs
hottest = myfile.loc[myfile['ANNUAL'].idxmax()]
coldest = myfile.loc[myfile['ANNUAL'].idxmin()]

print(f"\n Hottest YEAR: {int(hottest['YEAR'])} â {hottest['ANNUAL']}Â°C")
print(f" Coldest YEAR: {int(coldest['YEAR'])} â {coldest['ANNUAL']}Â°C")

#Seasonal Shift 
myfile['Winter_Avg'] = myfile[['JAN', 'FEB']].mean(axis=1)
myfile['Summer_Avg'] = myfile[['APR', 'MAY', 'JUN']].mean(axis=1)

# Calculate slope for both
winter_slope = np.polyfit(myfile['YEAR'], myfile['Winter_Avg'], 1)[0]
summer_slope = np.polyfit(myfile['YEAR'], myfile['Summer_Avg'], 1)[0]

print("\n Seasonal Shift:")
print(f"  - Winter warming rate: {winter_slope:.4f} Â°C/YEAR")
print(f"  - Summer warming rate: {summer_slope:.4f} Â°C/YEAR")

if winter_slope > summer_slope:
    print("  â Winters are warming faster than summers.")
elif summer_slope > winter_slope:
    print("  â Summers are intensifying more than winters.")
else:
    print("  â Both seasons are warming at similar rates.")
```
